```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
from skimage.filters import gaussian
%matplotlib notebook
from skimage.io import imread,imshow,imsave
```

```python
img=cv2.imread('Task01.jpg',0)
```

```python
gy=np.array([[1,-2,1],[0,0,0],[1,-2,1]])
gx=gy.T
```

```python
Gy=convolve2d(img,gy)
Gx=convolve2d(img,gx)
plt.figure()
plt.subplot(1,2,1)
plt.imshow(np.abs(Gy))
plt.subplot(1,2,2)
plt.imshow(np.abs(Gx))
plt.show()
```

```python
resultant=np.sqrt(Gy**2+Gx**2)
orientation=np.arctan2(Gy,Gx)
```

```python
plt.figure()
plt.subplot(1,2,1)
plt.imshow(np.abs(resultant))
plt.subplot(1,2,2)
plt.imshow(orientation,cmap=plt.cm.viridis)
plt.colorbar()
plt.show()
```
